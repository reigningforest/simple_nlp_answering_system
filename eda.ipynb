# Loading in the Data

```python
import pandas as pd
import os

def load_messages_to_dataframe(filename="data/all_messages.json"):
    """
    Reads a JSON file containing a list of message objects into a pandas DataFrame.

    The JSON file is expected to be in the 'records' format, which is a
    list of dictionaries.

    Args:
        filename (str): The path to the JSON file.

    Returns:
        pd.DataFrame: A pandas DataFrame containing the message data,
                      or None if the file is not found or is invalid.
    """
    if not os.path.exists(filename):
        print(f"Error: File not found at '{filename}'")
        return None

    try:
        # pd.read_json can directly handle a list of JSON objects
        df = pd.read_json(filename, orient='records')
        print(f"Successfully loaded {len(df)} messages into a DataFrame.")
        return df
    except ValueError as e:
        print(f"Error reading JSON file. It might not be in the expected format: {e}")
        return None
```

# Data Quality Analysis

Checking for anomalies and inconsistencies in the dataset.

## 1. Message Length Analysis

Identifying extremely short or long messages that might indicate data quality issues.

```python
# Calculate message lengths
df['message_length'] = df['message'].str.len()

# Summary statistics
print("Message Length Statistics:")
print(df['message_length'].describe())
print(f"\nMedian: {df['message_length'].median()}")

# Find outliers
q1 = df['message_length'].quantile(0.25)
q3 = df['message_length'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

extremely_short = df[df['message_length'] < 10]
extremely_long = df[df['message_length'] > upper_bound]

print(f"\n📊 Extremely short messages (<10 chars): {len(extremely_short)}")
if len(extremely_short) > 0:
    print("\nSample short messages:")
    print(extremely_short[['user_name', 'message', 'message_length']].head())

print(f"\n📊 Extremely long messages (>{upper_bound:.0f} chars): {len(extremely_long)}")
if len(extremely_long) > 0:
    print("\nSample long messages:")
    for idx, row in extremely_long.head().iterrows():
        print(f"\n{row['user_name']}: {row['message'][:100]}... (length: {row['message_length']})")

# Visualize distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['message_length'], bins=50, edgecolor='black', color='#4C72B0', alpha=0.7)
axes[0].axvline(lower_bound, color='red', linestyle='--', label=f'Lower Outlier Bound')
axes[0].axvline(upper_bound, color='red', linestyle='--', label=f'Upper Outlier Bound')
axes[0].set_title('Message Length Distribution', fontsize=14)
axes[0].set_xlabel('Message Length (characters)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].legend()

# Box plot
axes[1].boxplot(df['message_length'], vert=True)
axes[1].set_title('Message Length Box Plot', fontsize=14)
axes[1].set_ylabel('Message Length (characters)', fontsize=11)

plt.tight_layout()
plt.show()
```

## 2. Timestamp Validation

Checking for malformed or inconsistent timestamps.

```python
# Check for null/invalid timestamps
null_timestamps = df['timestamp'].isnull().sum()
print(f"📊 Null timestamps: {null_timestamps}")

# Check timestamp range
print(f"\n📅 Timestamp Range:")
print(f"Earliest: {df['timestamp'].min()}")
print(f"Latest: {df['timestamp'].max()}")
print(f"Span: {(df['timestamp'].max() - df['timestamp'].min()).days} days")

# Check for future timestamps (relative to data collection)
# Make pd.Timestamp.now() timezone-aware to match df timestamps
now_aware = pd.Timestamp.now(tz='UTC')
future_timestamps = df[df['timestamp'] > now_aware]
print(f"\n📊 Future timestamps: {len(future_timestamps)}")

# Check for timestamps before a reasonable date (e.g., year 2000)
cutoff_date = pd.Timestamp('2000-01-01', tz='UTC')
old_timestamps = df[df['timestamp'] < cutoff_date]
print(f"📊 Timestamps before 2000: {len(old_timestamps)}")

# Check timezone consistency (if timezone info exists)
if df['timestamp'].dt.tz is not None:
    print(f"\n🌍 Timezone: {df['timestamp'].dt.tz}")
else:
    print("\n🌍 No timezone information (timestamps are timezone-naive)")

# Look for duplicate timestamps within same user
print("\n📊 Checking for messages with identical timestamps per user...")
duplicate_times = df.groupby(['user_name', 'timestamp']).size().reset_index(name='count')
duplicate_times = duplicate_times[duplicate_times['count'] > 1]
print(f"Users with duplicate timestamps: {len(duplicate_times)}")
if len(duplicate_times) > 0:
    print("\nSample:")
    print(duplicate_times.head(10))
```

## 3. Time Between Messages Analysis

Analyzing message frequency and gaps per user.

```python
import numpy as np

# Calculate time between consecutive messages for each user
time_gaps = []

for user in df['user_name'].unique():
    user_df = df[df['user_name'] == user].sort_values('timestamp')
    user_df['time_diff'] = user_df['timestamp'].diff()
    
    # Convert to hours
    time_diffs_hours = user_df['time_diff'].dt.total_seconds() / 3600
    time_diffs_hours = time_diffs_hours.dropna()
    
    if len(time_diffs_hours) > 0:
        time_gaps.append({
            'user_name': user,
            'avg_hours': time_diffs_hours.mean(),
            'min_hours': time_diffs_hours.min(),
            'max_hours': time_diffs_hours.max(),
            'median_hours': time_diffs_hours.median(),
            'total_messages': len(user_df)
        })

time_gaps_df = pd.DataFrame(time_gaps).sort_values('avg_hours', ascending=False)

print("⏱️  Time Between Messages by User (in hours):\n")
print(time_gaps_df.to_string(index=False))

# Overall statistics
all_gaps = []
for user in df['user_name'].unique():
    user_df = df[df['user_name'] == user].sort_values('timestamp')
    user_df['time_diff'] = user_df['timestamp'].diff()
    gaps = user_df['time_diff'].dt.total_seconds() / 3600
    all_gaps.extend(gaps.dropna().tolist())

print(f"\n\n📊 Overall Time Gap Statistics:")
print(f"Average gap: {np.mean(all_gaps):.2f} hours")
print(f"Median gap: {np.median(all_gaps):.2f} hours")
print(f"Min gap: {np.min(all_gaps):.4f} hours ({np.min(all_gaps)*60:.2f} minutes)")
print(f"Max gap: {np.max(all_gaps):.2f} hours ({np.max(all_gaps)/24:.2f} days)")

# Visualize
fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(time_gaps_df['user_name'], time_gaps_df['avg_hours'], color='#4C72B0', alpha=0.7)
ax.set_xlabel('Average Hours Between Messages', fontsize=12)
ax.set_ylabel('User Name', fontsize=12)
ax.set_title('Average Time Between Messages by User', fontsize=14, pad=20)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. Duplicate Message Detection

Identifying exact duplicate messages that might indicate data quality issues.

```python
# Check for exact duplicate messages (same user, same message, same timestamp)
exact_duplicates = df[df.duplicated(subset=['user_name', 'message', 'timestamp'], keep=False)]
print(f"📊 Exact duplicate messages (same user, message, timestamp): {len(exact_duplicates)}")

if len(exact_duplicates) > 0:
    print("\nSample exact duplicates:")
    print(exact_duplicates[['user_name', 'timestamp', 'message']].head(10))

# Check for duplicate message content (same user, same message, different timestamp)
content_duplicates = df[df.duplicated(subset=['user_name', 'message'], keep=False)]
content_duplicates_count = len(content_duplicates) - len(exact_duplicates)  # Exclude exact duplicates
print(f"\n📊 Duplicate message content (same user, same text, different time): {content_duplicates_count}")

if content_duplicates_count > 0:
    # Show examples of repeated messages
    content_dup_groups = content_duplicates.groupby(['user_name', 'message']).size().reset_index(name='count')
    content_dup_groups = content_dup_groups[content_dup_groups['count'] > 1].sort_values('count', ascending=False)
    print(f"\nTop users with repeated messages:")
    print(content_dup_groups.head(10))
    
    # Show an example
    if len(content_dup_groups) > 0:
        example = content_dup_groups.iloc[0]
        print(f"\n📝 Example: {example['user_name']} sent the same message {example['count']} times:")
        print(f"Message: {example['message'][:100]}...")
        example_msgs = df[(df['user_name'] == example['user_name']) & (df['message'] == example['message'])]
        print("\nTimestamps:")
        for ts in example_msgs['timestamp'].head():
            print(f"  - {ts}")

# Check for ID duplicates
id_duplicates = df[df.duplicated(subset=['id'], keep=False)]
print(f"\n📊 Duplicate IDs: {len(id_duplicates)}")
if len(id_duplicates) > 0:
    print("⚠️  WARNING: Found messages with duplicate IDs!")
    print(id_duplicates[['id', 'user_name', 'timestamp', 'message']].head())

# Summary
print(f"\n\n📋 Duplicate Summary:")
print(f"Total messages: {len(df)}")
print(f"Unique messages: {df['message'].nunique()}")
print(f"Duplicate rate: {(1 - df['message'].nunique()/len(df))*100:.2f}%")
```
